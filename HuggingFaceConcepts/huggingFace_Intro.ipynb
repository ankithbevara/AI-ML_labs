Pipeline()- Makes it simpple to use any model from the **Hub**. It is the most powerful way to start using pre-trained Hugging Face Models.

```python
from transformers import pipeline

#load sentiment-analysis pipeline (sentiment - gives positive/ negative)
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
#Running Inference
result = sentiment_pipeline("I love teaching with generative AI")
print(result)

result1 = sentiment_pipeline("I dont know how to understand all the parameters in all the libraries")
print(result1)
```

```python
import torch
#Load text-generation pipeline with GPT-2
generator = pipeline("text-generation", model = "gpt2")
#model_id = "openai/gpt-oss-20b"
#generator = pipeline("text-generation", model=model_id, torch_dtype="auto", device_map="auto",)
#Generate text from a prompt
prompt = "in the future, AI-powered classrooms will"
output = generator(prompt, max_length=50, num_return_sequences=1)

print(output[0]["generated_text"])
```

#1. Text Classification

#2. Named Entity Recognition (NER)

```python
from transformers import pipeline

#specifying the inference task
ner_tagger = pipeline("ner")
```

```python
#Passing the input into the pipeline

ner_tagger(email)[:4]
```

```python
ner_tagger(product_review)[:4]
```

#3 Summarization

```python
summarizer(email)
```

```python
summarizer(product_review, max_length= 60)
```

#4: Question Answering

```python
from transformers import pipeline

#specify the inference task
reader = pipeline("question-answering")
```

```python
#passing the input to the pipeline
question = "when is the offer letter going to be shared?"

reader(question=question, context = email)
```

```python
#passing input to pipeline
question = "where did the customer buy the product?"

reader(question = question, context = product_review)
```

```python
question = "when is internship starting?"

reader(question = question, context = email)
```

#6 Text Generation

```python
from transformers import pipeline

#Specifying inference task
generator = pipeline("text-generation")
```

```python
prompt = f"""you are a helpful AI assistant.
Can you help me write the response for the following product review:
#### PRODUCT REVIEW ###
{product_review}

### CUSTOMER SERVICE RESPONSE ###
Dear Customer, I am sorry to hear this. please be assured that
"""

print(prompt)
```

```python
response = generator(prompt, max_length = 200)

response
```

```python
print(response[0]["generated_text"])
```

#NLP Task 8- Feature Extraction

```python
from transformers import pipeline

#Load the feature extractio pipeline and model
pipe = pipeline('feature-extraction')
```

```python
sentence= """
AI encompasses a range of technologies, including machine learning, natural language processing,
robotics and more
"""

#Get Embeddings
embedding = pipe(sentence)
```

```python
#Output is a list with one item per sentence

sentence_embeddings = embedding[0][0][0:10]
print(sentence_embeddings)
```
