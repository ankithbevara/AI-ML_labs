**USING GEMINI MODEL TO SUMMARIZE TEXT**

```python
#import google gemini
import google.generativeai as genai
```

```python
#Code to access api key of gemini
f = open("gemini_key.txt")
key = f.read()
genai.configure(api_key=key)
```

```python
#using gemini model to summarize the text
model = genai.GenerativeModel("gemini-2.0-flash")

#Read input text data from a file
file_name = open("summarizedtext.txt")
text = file_name.read()

#create a prompt
prompt = f"summarize the following text in a concise paragraph:\n\n{text}"
response = model.generate_content(prompt)
print(response.text)
```

What if we want to save this summarized text in new file?

```python
new_file = open("sum_summarizedtext.txt", "w")
new_file.write(response.text)
new_file.close()
```
