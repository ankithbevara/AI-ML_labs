```python
import google.generativeai as genai

API_KEY=open("gemini_key.txt", "r").read()

#Configure GEMINI API KEY
genai.configure(api_key = API_KEY)

#Initialize the Gemini Model
model = genai.GenerativeModel("gemini-2.0-flash")
```

```python
#Prompt Builder for different Writing Styles
def build_prompt(style, topic):
  style= style.lower()
  if style == "story":
    return f"Write a short story about:{topic}"
  elif style == "poem":
    return f"Write a poem based on theme:{topic}"
  elif style == "blog":
    return f"write a blog post about:{topic}"
  elif style == "script":
    return f"write a script based on: {topic}"
  else:
    return f"write as a generative AI assistant: {topic}"
```

```python
#Generating reponse using Gemini
def generate_response(style, topic):
  prompt = build_prompt(style, topic)
  response = model.generate_content(prompt)
  return response.text
```

```python
#Example 1
print(generate_response("story", "how did india got independence"))
```

```python
#Example 1
print(generate_response("poem", "rain and water"))
```
