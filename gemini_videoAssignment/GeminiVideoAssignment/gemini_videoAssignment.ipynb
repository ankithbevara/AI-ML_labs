```python
#Importing google gemini AI
import google.generativeai as genai
```

```python
#Code to access api key of gemini
f = open("gemini_key.txt")
key = f.read()
genai.configure(api_key=key)
```

**UPLOADING FILE**

```python
import time

video_file = genai.upload_file(path="/content/video.mp4")

#Waiting for the video to upload before calling model
while True:
    file_info = genai.get_file(video_file.name)
    if file_info.state.name == "ACTIVE":
        break
    time.sleep(2)   # wait 2 seconds
```

```python
#Choose Model:
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
prompt = """
Summarize this video in:
- Title of the Video
- Key highlights/important events
- A short paragraph Summary
"""

response = model.generate_content(
    [video_file, prompt],
    request_options={"timeout": 600}
)

print(response.text)
```
